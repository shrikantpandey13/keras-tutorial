```python
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dropout,Dense,Activation
from keras.optimizers import Adam,SGD,RMSprop
from keras.utils import np_utils
import matplotlib.pyplot as plt
import numpy as np
```

```python
BATCH_SIZE= 128
VALIDATION_SPLIT = 0.33
NB_CLASSES = 10
VERBOSE =1

```

```python
# DIVDIE THE DATA TRAIN AN DTEST
(X_train,y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train = X_train.reshape(60000,784)
X_test = X_test.reshape(10000,784)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# normalize the data
X_train /= 255
X_test /= 255

print("shape of train data",X_train.shape)
print("test of tests data",X_test.shape)
```

```python
## connvert classes to categrocal data
y_train = np_utils.to_categorical(y_train,NB_CLASSES)
y_test = np_utils.to_categorical(y_test,NB_CLASSES)
```

```python
accuracy = []
for OPTIMIZERS in [Adam(),SGD(),RMSprop()]:
    for NB_EPOCH in[5,10,20]:
        for Multiple_Layers in [64,128,256]:
            model = Sequential()
            model.add(Dense(units=Multiple_Layers,input_shape=(784,)))
            model.add(Activation('relu'))
            model.add(Dense(units=Multiple_Layers))
            model.add(Activation('relu'))
            model.add(Dense(units=NB_CLASSES))
            model.add(Activation('softmax'))
            model.summary()
            
            model.compile(loss='categorical_crossentropy',optimizer = OPTIMIZERS,metrics=['accuracy'])
            model.fit(X_train,y_train,batch_size=BATCH_SIZE,epochs=NB_EPOCH,validation_split=VALIDATION_SPLIT)
            
            score = model.evaluate(X_test,y_test,verbose=VERBOSE)
            print("Optimizers:",OPTIMIZERS)
            print("epoch size:",NB_EPOCH)
            print("Numbers of layers:",Multiple_Layers)
            print("test score",score[0])
            print("test accuracy:",score[1])
            accuracy.append(score[1])
            
print(accuracy)

            
            
            
            
```
